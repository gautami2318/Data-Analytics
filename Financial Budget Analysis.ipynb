```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV
data = pd.read_csv('India_budget_2021.csv')

# Remove rows with missing values in key columns
data = data.dropna(subset=['Department /Ministry', 'Fund allotted(in ₹crores)'])

# Remove commas, ₹ symbols, or anything that's not a number
data['Fund allotted(in ₹crores)'] = (
    data['Fund allotted(in ₹crores)']
    .astype(str)
    .str.replace(',', '', regex=False)
    .str.extract(r'([\d.]+)')[0]  # Extract only the numeric part
)

# Convert to float
data['Fund allotted(in ₹crores)'] = pd.to_numeric(data['Fund allotted(in ₹crores)'], errors='coerce')

# Drop the GRAND TOTAL row if it's still there
data = data[data['Department /Ministry'].str.upper() != 'GRAND TOTAL']

# Plot
plt.figure(figsize=(14, 8))
data.plot.bar(x='Department /Ministry', y='Fund allotted(in ₹crores)', legend=False)
plt.ylabel('Fund allotted (₹ Crores)')
plt.title('Budget Allocation by Department / Ministry (2021)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
data = data.sort_values(by='Fund allotted(in ₹crores)', ascending=False)
```

```python

df = data["Fund allotted(in ₹crores)"]
labels = data["Department /Ministry"]
plt.figure(figsize=(7,7))
plt.pie(df, labels=labels, autopct='%1.1f%%', startangle=90, pctdistance=0.85, shadow =True)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Distribution of The Budget", fontsize=20)
plt.show()
```
